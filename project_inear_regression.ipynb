```python
# import libraries
import numpy as np
import pandas as pd 
```

```python
df=pd.read_csv("HousePricePrediction.csv")
df
```

```python
# Feature engineering
df.dropna(subset=['SalePrice'],inplace=True)
x=df.drop(["MSZoning","Exterior1st","BldgType","SalePrice"], axis=1)
y=df["SalePrice"]
```

```python
# handle missing values in sale price target column
# df.dropna(subset=['SalePrice'])
```

```python


print(x.columns)
```

```python
# encoding
x["LotConfig"]=x["LotConfig"].map({
    "Inside": 0,
    "Corner": 1,
    "CulDSac": 2,
    "FR2": 3,
    "FR3": 4
})
```

```python
x
```

```python
# test and training
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
    
```

```python
x_train.head()
```

```python
# Train the model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
# prdict values
y_predict=model.predict(x_test)
y_predict
```

```python
# evaluate the model
from sklearn.metrics import r2_score
r2=r2_score(y_test, y_predict)
print(r2)
```
